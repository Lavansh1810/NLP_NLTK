```python
# POS Tagging

# A POS tag (or part-of-speech tag) is a special label assigned to each token (word) in a text corpus to indicate the part of speech and often also other grammatical categories such as tense, number (plural/singular), case etc. 

# Some NLTK POS tagging examples are: CC, CD, EX, JJ, MD, NNP, PDT, PRP$, TO, etc. 
from nltk.tokenize import word_tokenize

text = "My name is Lavansh Arora. This is my NLP assignment."

tokens_in_text = word_tokenize(text)

nltk.pos_tag(tokens_in_text)
```

```python
#Possible tagsets in nltk are
# nltk.download('tagsets')
nltk.help.upenn_tagset()
```

```python
# Semantic tagging

# Semantic tagging is the process of connecting an element from an ontology with a database or document, such as a website or computer file. This form of text annotation allows us to efficiently describe key components of a data structure to facilitate better retrieval of them later on or understand how they relate with other resources in the same ontology.

# Semantic Evaluate Package 

from nltk.sem.evaluate import Assignment

dom = set(['lavansh', 'arora', '70'])

g3 = Assignment(dom, [('n', 'lavansh'), ('s', 'arora')])

g3 == {'n': 'lavansh', 's': 'arora'}

g3.add('w', '70')

print(g3)

# With no arguments, purge() is equivalent to clear() on a dictionary

g3.purge()

g3
```

###Create a text corpus with a minimum of 200 words (unique content). Implement the following text processing.

```python
text = '''I love to eat food. I have been to many countries around the world have tried ample of food varieties. Despite of all this, my favourite category of food is Indian food. The types of cuisine available, the flavours, and the serving customs vary according to region in India. Preservatives are typically not used in Indian cuisine. You will so obtain all the nutrients you require. Additionally, yoghurt, which is fermented milk and is used in Indian cuisine, aids in keeping good bacteria. The majority of the techniques used to prepare Indian food maintain the nutrients included in the fresh veggies utilised. Indian cuisine is diverse and offers a lot of choice. India's states each have their own distinct cuisine that embodies their traditions and culture. Initially, Kashmiri cuisines in northern India have significant Central Asian influences. In Kashmir, nearly every meal is built on rice, which is grown in abundance across the stunning valley. Another delectable dish created here is the "Saag," which is made with a green leafy vegetable called the "Hak." However, places like Punjab, Haryana, and Uttar Pradesh have high chapatis consumption as a staple cuisine. Again, a variety of flours, including wheat, rice, maida, besan, and others, are used to make these chapatis. The desert cuisine of western India is renowned for its distinct flavour and variety of foods. The two states that best exemplify the desert character of Indian cuisine are Rajasthan and Gujarat. The Bengali and Assamese culinary traditions are evident throughout eastern India. The delicious dish of rice and fish is the main dish of Bengalis. Typically, Bengalis like eating a variety of fish. Since most of the states in southern India have coastal kitchens, they make extensive use of spices, seafood, and coconuts. The tamarind fruit is commonly used in Tamil Nadu cuisine to give dishes a sour flavour.'''
```

```python
# Word segmentation

print("The length of word tokens is : ", len(word_tokenize(text)))

print("The list of segmented word is : ", (word_tokenize(text)))
```

```python
# Sentence segmentation

print("The length of segmented scentence list is : ", len(sent_tokenize(text)))

print("The list of segmented scentence is : ", (sent_tokenize(text)))
```

```python
# Convert to Lowercase

lowercase = text.lower()
print(lowercase)
```

```python
# Stop words removal

import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords

stopwords = stopwords.words('english')

answer = []

word_tokens = word_tokenize(lowercase)

for word in word_tokens:
  if(word not in stopwords):
    answer.append(word)

print("Corpous after stop words rwmoval is : ",answer)
```

```python
# Stemming

from nltk.stem import PorterStemmer
  
ps = PorterStemmer()

word_tokens = word_tokenize(text)

for w in word_tokens:
    print(w, " : ", ps.stem(w))
```

```python
# Lemmatization

from nltk.stem import WordNetLemmatizer
# nltk.download('wordnet')
# nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

word_tokens = word_tokenize(text)

for w in word_tokens:
    print(w, " : ", lemmatizer.lemmatize(w))
```

```python
# Part of speech tagger
# nltk.download('averaged_perceptron_tagger')

word_tokens = word_tokenize(text)

nltk.pos_tag(word_tokens)
```
